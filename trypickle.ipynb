```python
import pandas as pd
import pickle
import numpy as np
```

```python
x_train = pd.read_csv("train_data.csv")
x_test = pd.read_csv("test_data.csv")
```

```python
x_train
```

```python
x_train.drop('Unnamed: 0', axis=1, inplace=True)
x_test.drop('Unnamed: 0', axis=1, inplace=True)
```

# Open Model

### Using Train Data

```python
pickled_model = pickle.load(open('Churn_rf.pkl', 'rb'))
y_predict = pickled_model.predict(x_train)
result = pd.DataFrame(y_predict, columns=['Churn'])
result.value_counts()
```

### Using Test Data

```python
pickled_model = pickle.load(open('Churn_rf.pkl', 'rb'))
y_predict = pickled_model.predict(x_test)
result = pd.DataFrame(y_predict, columns=['Churn'])
result.value_counts()
```
